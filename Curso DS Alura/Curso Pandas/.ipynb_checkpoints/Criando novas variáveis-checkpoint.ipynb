## Criando novas variaveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

```python
#Criando coluna e atribuindo valores com algebra
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head(10)
```

```python
dados['Valor Bruto m2'] = dados['Valor Bruto'] / dados['Area']
```

```python
dados.head(10)
```

```python
# retorna os tipos unicos(as possibilidades) da coluna
dados['Tipo'].unique()
```

```python
#Criando lista com tipos de casa
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
#Criando coluna com a simplificação de tipo de moradia. Utilizando lambda
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
#Visualizando a tabela com as novas colunas
selecao = dados['Tipo'] == 'Casa de Condomínio'
dados[selecao]
```

## Excluindo Variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
```

```python
dados_aux.head(10)
```

```python
#Deletar coluna
del dados_aux['Valor Bruto']
```

```python
dados_aux
```

```python
#Tambem deleta coluna
dados_aux.pop('Valor Bruto m2')
```

```python
dados_aux
```

```python
#Deleta linhas ou colunas(axis 0 ou 1)
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True)
```

```python
dados.head(10)
```

```python
dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```
